```python
import pandas as pd

# Assuming kinematic_df is your DataFrame and it's already loaded

# Calculate the average pitch speed
average_pitch_speed = kinematic_df['pitch_speed_mph'].mean()

# Function to categorize pitch speeds, used within the value_counts() method
def categorize_pitch_speed(speed):
    if speed >= 90:
        return '90+ mph'
    elif 85 <= speed < 90:
        return '85-90 mph'
    elif 80 <= speed < 85:
        return '80-85 mph'
    else:
        return 'Less than 80 mph'

# Count the number of pitches in each speed category without altering the original DataFrame
pitch_speed_counts = kinematic_df['pitch_speed_mph'].apply(categorize_pitch_speed).value_counts()

# Output the average pitch speed and the counts
print("Average Pitch Speed (mph):", average_pitch_speed)
print("\nNumber of Pitches by Speed Category:")
print(pitch_speed_counts)


kinematic_df.describe()
```

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np

# Splitting the data into features (X) and target (y)
X_kinematic = kinematic_df.drop('pitch_speed_mph', axis=1)
y_kinematic = kinematic_df['pitch_speed_mph']

# Splitting the data into training and test sets
X_train_km, X_test_km, y_train_km, y_test_km = train_test_split(X_kinematic, y_kinematic, test_size=0.2, random_state=42)

# Initializing the scaler
scaler = StandardScaler()

# Fitting the scaler on the training data and transforming both training and test sets
X_train_km_scaled = scaler.fit_transform(X_train_km)
X_test_km_scaled = scaler.transform(X_test_km)

# Linear regression for kinetic variables using scaled data
lin_reg_km = LinearRegression()
lin_reg_km.fit(X_train_km_scaled, y_train_km)

# Predictions on the test set
y_pred_km_test = lin_reg_km.predict(X_test_km_scaled)

# Predictions on the training set for comparison
y_pred_km_train = lin_reg_km.predict(X_train_km_scaled)

# Evaluate the model on the test set
print("Kinematic Linear Regression (Test Set) R-squared:", r2_score(y_test_km, y_pred_km_test))
print("Kinematic Linear Regression (Test Set) RMSE:", np.sqrt(mean_squared_error(y_test_km, y_pred_km_test)))

# Evaluate the model on the training set
print("Kinematic Linear Regression (Training Set) R-squared:", r2_score(y_train_km, y_pred_km_train))
print("Kinematic Linear Regression (Training Set) RMSE:", np.sqrt(mean_squared_error(y_train_km, y_pred_km_train)))
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

# Assuming kinematic_df is already defined and includes 'pitch_speed_mph'
# Splitting the data into features (X) and target (y)
X_kinematic = kinematic_df.drop('pitch_speed_mph', axis=1)
y_kinematic = kinematic_df['pitch_speed_mph']

# Split the data into training and test sets
X_train_km, X_test_km, y_train_km, y_test_km = train_test_split(X_kinematic, y_kinematic, test_size=0.2, random_state=42)

# Initialize and train the Random Forest regressor
rf_regressor = RandomForestRegressor(n_estimators=100, random_state=42)
rf_regressor.fit(X_train_km, y_train_km)

# Make predictions on the test set and the training set
y_pred_km_test = rf_regressor.predict(X_test_km)
y_pred_km_train = rf_regressor.predict(X_train_km)

# Evaluate the model on the test set
r2_test = r2_score(y_test_km, y_pred_km_test)
rmse_test = np.sqrt(mean_squared_error(y_test_km, y_pred_km_test))
print(f"Test Set - R-squared: {r2_test:.2f}, RMSE: {rmse_test:.2f}")

# Evaluate the model on the training set
r2_train = r2_score(y_train_km, y_pred_km_train)
rmse_train = np.sqrt(mean_squared_error(y_train_km, y_pred_km_train))
print(f"Training Set - R-squared: {r2_train:.2f}, RMSE: {rmse_train:.2f}")

# Feature importance - focusing on the top 10 features
top_n = 10
feature_importances = pd.Series(rf_regressor.feature_importances_, index=X_kinematic.columns).sort_values(ascending=False).head(top_n)

# Plotting feature importances
plt.figure(figsize=(10, 8))
feature_importances.plot(kind='barh', color='#2ca02c')
plt.title('Top 10 Feature Importances in Random Forest Model')
plt.xlabel('Importance')
plt.ylabel('Features')
plt.gca().invert_yaxis()
plt.show()
```

```python
import seaborn as sns
sns.set_palette("pastel")

# Create a DataFrame for seaborn
data = {'Actual Speed': y_test_km, 'Predicted Speed': y_pred_km_test}
df = pd.DataFrame(data)

# Scatter plot with X marker for blue dots
plt.figure(figsize=(5, 5))
sns.scatterplot(x='Actual Speed', y='Predicted Speed', data=df, marker='X', color='blue', alpha=0.7, label='Predictions')

# Add the line y = x
plt.plot([min(df['Actual Speed']), max(df['Actual Speed'])], [min(df['Actual Speed']), max(df['Actual Speed'])], linestyle='--', color='red', label='Ideal Prediction Line')

plt.xlabel('Actual Speed')
plt.ylabel('Predicted Speed')
plt.title('Forest Regression: Actual vs Predicted Speeds')
plt.legend()  # Add legend
plt.grid(True)
plt.show()
```
